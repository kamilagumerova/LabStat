```python
import numpy as np
import pandas as pd
```

## 1. Загрузка

```python
df_start = pd.read_csv('auto.csv')
df_start
```

```python
df = df_start[['symboling', 'normalized-losses', 'wheel-base', 'length',
               'width', 'height', 'curb-weight', 'engine-size', 'bore',
               'stroke', 'compression-ratio', 'peak-rpm', 'city-mpg',
               'highway-mpg', 'price']]
df
```

## 2. Основные статистики
    

```python
from statistics import mode

df.agg(['min', 'max', np.mean, np.median, np.std, mode])
```

## 3. Матрица корреляции

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 10))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

## 4. Распределение Price и skewness, и kurtosis

```python
plt.hist(df['price'])

plt.show()

print('Kurtosis =', df['price'].kurtosis())
print('Skewness =', df['price'].skew())
```

## 5. Визуализация взаимосвязи между price и engine-size, и между curb-weight и highway-mpg

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 7))

axes[0].scatter(df['price'], df['engine-size'], s=20)
axes[0].set_title('Связь price и engine-size')
axes[0].set_ylabel('price')
axes[0].set_xlabel('engine-size')

axes[1].scatter(df['curb-weight'], df['highway-mpg'], s=20)
axes[1].set_title('Связь curb-weight и highway-mpg')
axes[1].set_ylabel('curb-weight')
axes[1].set_xlabel('highway-mpg')

plt.show()
```

## 6. Гипотезы и их проверка

```python
from scipy.stats import ttest_ind
```

Т-тест можно использовать только для данных с приблизительным нормальным распределением

```python
plt.hist(df['stroke'])
```

```python
plt.hist(df['width'])
```

**Нулевая гипотеза:** ход поршня у машин с 2-мя дверями и с 4-мя не отличаются, коэффициент p для отказа от гипотезы должен быть меньше 0.05

```python
ttest_ind(df_start[df_start['num-of-doors'] == 'two']['stroke'],
          df_start[df_start['num-of-doors'] == 'four']['stroke'])
```

Коэффициент p больше 0.05 - гипотезу не отклоняем

**Нулевая гипотеза:** ширина не зависит от того седан это или хэтчбек, коэффициент p для отказа от гипотезы должен быть меньше 0.05

```python
ttest_ind(df_start[df_start['body-style'] == 'sedan']['width'],
          df_start[df_start['body-style'] == 'hatchback']['width'])
```

Коэффициент p меньше 0.05 - гипотезу отклоняем
